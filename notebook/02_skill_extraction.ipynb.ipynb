```python
import pandas as pd

jobs_clean = pd.read_csv(
    "cleaned_jobs.csv"
)
```

```python
skills_dictionary = [
    'sql',
    'excel',
    'python',
    'power bi',
    'tableau',
    'aws',
    'azure',
    'spark',
    'snowflake',
    'etl',
    'statistics',
    'machine learning',
    'data visualization'
]
```

```python
def extract_skills(description):

    description = str(description).lower()

    found_skills = []

    for skill in skills_dictionary:

        if skill in description:

            found_skills.append(skill)

    return found_skills
```

```python
jobs_clean['technical_skills'] = (
    jobs_clean['job_description']
    .apply(extract_skills)
)
```

```python
jobs_clean[
['job_title','technical_skills']
].head(10)
```

```python
jobs_clean['technical_skills_text'] = (
    jobs_clean['technical_skills']
    .apply(lambda x: ', '.join(x))
)
```

```python
from collections import Counter
```

```python
all_skills = []

for skills in jobs_clean[
    'technical_skills'
]:
    all_skills.extend(skills)
```

```python
skill_counts = Counter(
    all_skills
)
```

```python
skill_demand = pd.DataFrame(
    skill_counts.items(),
    columns=[
        'skill',
        'count'
    ]
)
```

```python
skill_demand = skill_demand.sort_values(
    by='count',
    ascending=False
)
```

```python
skill_demand
```

```python
skill_demand.to_csv(
    "skill_demand.csv",
    index=False
)
```
