```python
import pandas as pd
import numpy as np
```

```python
postings = pd.read_csv("postings.csv")
```

```python
postings.head()
```

```python
postings.shape
```

```python
postings.columns.tolist()
```

```python
keywords = [
    'data analyst',
    'business analyst',
    'data scientist',
    'data engineer',
    'bi analyst'
]
```

```python
jobs = postings[
    postings['title']
    .str.lower()
    .str.contains('|'.join(keywords), na=False)
]
```

```python
jobs.shape
```

```python
jobs['title'].value_counts().head(20)
```

```python
keep_cols = [
    'job_id',
    'company_name',
    'title',
    'description',
    'location',
    'formatted_work_type',
    'remote_allowed',
    'min_salary',
    'max_salary',
    'applies'
]
```

```python
jobs_clean = jobs[keep_cols].copy()
```

```python
jobs_clean.head()
```

```python
jobs_clean.rename(
    columns={
        'company_name':'company',
        'title':'job_title',
        'description':'job_description',
        'formatted_work_type':'job_type'
    },
    inplace=True
)
```

```python
jobs_clean.columns
```

```python
jobs_clean['avg_salary'] = (
    jobs_clean['min_salary'] +
    jobs_clean['max_salary']
)/2
```

```python
jobs_clean[
['min_salary','max_salary','avg_salary']
].head()
```

```python
jobs_clean.duplicated().sum()
```

```python
jobs_clean.isnull().sum()
```

```python
jobs_clean = jobs_clean.dropna(
    subset=['job_description']
)
```

```python
jobs_clean['company'] = jobs_clean[
    'company'
].fillna('Unknown')
```

```python
jobs_clean['location'] = jobs_clean[
    'location'
].fillna('Unknown')
```

```python
jobs_clean.info()
```

```python
jobs_clean.to_csv(
    "cleaned_jobs.csv",
    index=False
)
```
